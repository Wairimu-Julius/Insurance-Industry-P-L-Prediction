# Applying Machine Learning On the Data Set

```python
#1.Decision Tree Classifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(['Target','Profit','Profit transferred to P&L','Company','Year'],axis=1)
y=data['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions=dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,predictions))

print(classification_report(y_test,predictions))
```

```python
#2.Random Forest Classifier Model
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=300)
rfc.fit(X_train,y_train)
```

```python
predictions=rfc.predict(X_test)
print(confusion_matrix(y_test,predictions))

print(classification_report(y_test,predictions))
```

```python
#3.Logistic Model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,y_train)
```

```python
predictions=logistic_regression.predict(X_test)

print(confusion_matrix(y_test,predictions))

print(classification_report(y_test,predictions))
```

Understanding Features Importance and their contribution to the model

```python
import shap
```

```python
shap.initjs()
```

```python
explainer = shap.LinearExplainer(logistic_regression, X_train, feature_dependence="independent")
shap_values = explainer.shap_values(X_test)
```

```python
shap.summary_plot(shap_values, X_test, plot_type="bar")
```

```python
shap.summary_plot(shap_values, X_test)
```

Understanding how the model features contributed to the outcome of a specific Insurer to either Loss/Profit Making

```python
# Profit Making
shap.force_plot(explainer.expected_value, shap_values[0,:], X_test.iloc[0,:])
```

```python
#Loss Making
shap.force_plot(explainer.expected_value, shap_values[1,:], X_test.iloc[1,:])
```

```python
shap.force_plot(explainer.expected_value, shap_values, X_test)
```

```python
#shap.plots.waterfall(shap_values[28])
shap.waterfall_plot(explainer.expected_value, shap_values[28,:], feature_names=X.columns.values, max_display=20, show=True)
```
